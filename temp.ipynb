```python
import matplotlib.pyplot as plt
import torch
import numpy as np

subnet_weights_classes = torch.load('/home/local/ASURITE/rgoel15/ICML_VLM/subnet_weights.pt')
subnet_weights_classes = subnet_weights_classes.detach().cpu().float().numpy()
print(subnet_weights_classes.shape)
```

```python
# Plotting each class with histograms for IB Loss and Subnet Weights
for i in range(5):
    # Subplot 2: Histogram for Subnet Weights
    plt.subplot(1, 2, 2)
    plt.hist(subnet_weights_classes[:, i], bins=30, alpha=0.7, color='green')
    plt.title(f"Class {i}: Subnet Weights Distribution")
    plt.xlabel(f"Subnet Weights for Class {i}")
    plt.ylabel("Frequency")

    plt.tight_layout()
    plt.show()
```
